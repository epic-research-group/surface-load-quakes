```python
# Goal is to write a new CSV file with columns:
# id,latitude,longitude,time,magnitude
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from global_land_mask import globe
```

```python
# Load datasets
earthquake_catalog = pd.read_csv('eq_200204_202102.csv')
earthquake_catalog["time"] = earthquake_catalog["time"].astype("datetime64")
```

```python
earthquake_catalog_mini = earthquake_catalog.drop(columns=['depth', 'magType','nst','gap','dmin','net','horizontalError',
                                                     'updated','rms','place','type','depthError','magError','status',
                                                     'locationSource','magSource','magNst','id'])
earthquake_catalog_mini = earthquake_catalog_mini[['latitude', 'longitude', 'time', 'mag']]
earthquake_catalog_mini = earthquake_catalog_mini.rename(columns={"mag": "magnitude"})
earthquake_catalog_mini.to_csv (r'eq_200204_202102_mini.csv', index = True, header=True)
```
